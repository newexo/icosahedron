```python
from icosahedron import directories
from icosahedron.generate import generate_code_context
```

```python
def character_markdown():
    code = generate_code_context.code_context(
        "character.py", base_dir=directories.code("d20_rules")
    )
    unittest = generate_code_context.unittest_code_context("test_character.py")

    return "\n\n".join([code, unittest])
```

```python
print(character_markdown())
```

```python
with open(directories.data("generated_context/character.md"), "w") as f:
    f.write(character_markdown())
```
